```python
#Import the required Libraries
from tkinter import *
from PIL import Image,ImageTk
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from random import sample


#Create an instance of tkinter frame
win = Tk()

#Set the geometry of tkinter frame
win.geometry("750x750")
df = pd.read_excel("C:\\Users\\Ishaan\\Desktop\\Crop_recommendation.xlsx")
l=set()
for i in range(len(df)):
    x=df.loc[i,'label']
    l.add(x)
dfs=df.loc[df['label'] == sample(l,1)[0]]
fg=sample(ls,1)
ll=sns.jointplot(x='humidity',y='temperature',data=dfs,kind='hex',color=fg[0])
plt.title(' Rice')
plt.savefig('1.png')
plt.annotate("User data", (20,20) )

#Create a canvas
canvas= Canvas(win, width= 600, height= 400)
canvas.pack()

#Load an image in the script
img= ImageTk.PhotoImage(Image.open("C:\\Users\\Ishaan\\1.png"))

#Add image to the Canvas Items
canvas.create_image(10,10,anchor=NW,image=img)

win.mainloop()
os.remove("C:\\Users\\Ishaan\\1.png")
```
